# BIAS AND VARIENCE -- TRADE OFF

```python
import numpy as np
import matplotlib.pyplot as pt
t=np.linspace(0,1,12)
h=np.random.randint(12,30,12)
pt.scatter(h,t)
pt.show()
```

```python
p2=np.polyfit(t,h,1)  
pt.scatter(t,h)                            # Θ0 + Θ1(x)
pt.plot(t,np.polyval(p2,t),c='r')          # degree of polynomial=1,high training error,high test error
pt.plot(t,np.polyval(p2,t),c='r')          # bias-high, variance-high,under-fitting
pt.xlabel('t(s)')
pt.ylabel('h(m)')
pt.title('Polynomial Model is underfit')
pt.show()
```

```python
p3=np.polyfit(t,h,2)  
pt.scatter(t,h)                                  # Θ0 + Θ1(x) + Θ2(x)2 + Θ3(x)2
pt.plot(t,np.polyval(p3,t),c='r')                #degree of polynomial=2,low training error,low test error
pt.xlabel('t(s)')                                #bias-low,varience-low,generalized/robust apprepriate fitting
pt.ylabel('h(m)')
pt.title('Linear Model is underfit')
pt.show()
```

```python
p2=np.polyfit(t,h,15)  
pt.scatter(t,h)                                    # Θ0 + Θ1(x) + Θ2(x)2 + Θ3(x)2+....+ Θ14(x)14
pt.plot(t,np.polyval(p2,t),c='r')                  #degree of polynomial=14,low training error,high test error
pt.xlabel('t(s)')                                  # bias-low,variance-high,overfitting
pt.ylabel('h(m)')
pt.title('Polynomial Model is overfit')
pt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Generate some data for classification
np.random.seed(0)
n_samples = 50
X = np.random.randn(n_samples, 2)
y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)

# Fit a logistic regression model with low regularization
clf = LogisticRegression(C=1000, penalty='l2')
clf.fit(X, y)

# Plot the decision boundary
plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
xx, yy = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[0.5], colors='k')
plt.title("Underfitting Classification Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
np.random.seed(0)
X = np.linspace(0, 1, 20)
y = np.sin(2*np.pi*X) + np.random.randn(20)*0.2

# Define a function to generate polynomial features
def polynomial_features(X, degree):
    n_samples = X.shape[0]
    X_poly = np.zeros((n_samples, degree+1))
    for d in range(degree+1):
        X_poly[:,d] = X**d
    return X_poly

# Fit a polynomial model of degree 10 to the data
degree = 10
X_poly = polynomial_features(X, degree)
w = np.linalg.solve(X_poly.T.dot(X_poly), X_poly.T.dot(y))
y_pred = X_poly.dot(w)

# Plot the data and the model prediction
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Model')
plt.title(f'Overfitting (Degree={degree})')
plt.legend()
plt.show()
```
