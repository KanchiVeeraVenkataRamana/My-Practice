```python
from sklearn import datasets
df=datasets.load_iris()
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score,KFold,StratifiedKFold
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.data,df.target,test_size=0.2,random_state=3)
```

```python
x=x_train
print(x.shape)
print(x.size)
```

```python
print(x_test.shape)
print(x_test.size)
```

```python
y=y_train
print(y.shape)
print(y.size)
```

```python
print(y_test.shape)
print(y_test.size)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
ml=[LogisticRegression(),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
def models_train_test():
    for model in ml:
        model.fit(x_train,y_train)
        accuracy=metrics.accuracy_score(y_test,model.predict(x_test))
        print('Accuracy Score of the',model,'=',accuracy)
        
models_train_test()
```

```python
cv_score_lr=cross_val_score(LogisticRegression(),x,y,cv=5)
print(cv_score_lr)
```

```python
mean_accuracy_lr=sum(cv_score_lr)/len(cv_score_lr)
mean_accuracy_lr=mean_accuracy_lr*100
mean_accuracy_lr=round(mean_accuracy_lr,2)
print(mean_accuracy_lr)
```

```python
mean_accuracy_lr
```

```python
#1.Cross_validation
def models_cross_val():
    for model in ml:
        cv_score=cross_val_score(model,x,y,scoring='accuracy',cv=5,n_jobs=1)
        mean_accuracy=sum(cv_score)/len(cv_score)
        mean_accuracy=mean_accuracy*100
        #mean_accuracy=round(mean_accuracy_socre,2)
        print('Cross Validation accuries for',model,'=',cv_score)
        print('Accuracy % of the',model,':',mean_accuracy)
        print('_________________________________________________________________________________________________________')
    
models_cross_val()
```

```python
#1.Cross_validation
def models_cross_val():
    for model in ml:
         #scoring = {'prec_macro': 'precision_macro','rec_macro'}
        cv_score=cross_val_score(model,x,y, scoring='r2',cv=5)
        mean_accuracy=sum(cv_score)/len(cv_score)
        mean_accuracy=mean_accuracy*100
        #mean_accuracy=round(mean_accuracy_socre,2)
        print('Cross Validation accuries for',model,'=',cv_score)
        print('Accuracy % of the',model,':',mean_accuracy)
        print('_________________________________________________________________________________________________________')
    
models_cross_val()
```

```python
#2.KFold Cross_validation
def models_cross_val():
    for model in ml:
        cv_score=cross_val_score(model,x,y,cv=KFold(n_splits=5),scoring='accuracy',n_jobs=1)
        mean_accuracy=sum(cv_score)/len(cv_score)
        mean_accuracy=mean_accuracy*100
        mean_accuracy=round(mean_accuracy,2)
        print('Cross Validation accuries for',model,'=',cv_score)
        print('Accuracy % of the',model,':',mean_accuracy)
        print('_________________________________________________________________________________________________________')
    
models_cross_val()
```

```python
#1.GridSearchSearchCV
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
model=SVC()
parameters={'kernel':['linear','poly','sigmoid','rbf'],'C':[1,5,10,20]}
```

```python
#cls=GridSearchCV(cv=5,estimator=SVC(),param_grid={'C':[1,2,10,20],'kernel':['linear','poly','rbf','sigmoid']})
```

```python
cls=GridSearchCV(model,parameters,cv=5)
```

```python
cls.fit(x,y)
```

```python
cls.cv_results_
```

```python
cls.best_params_
```

```python
cls.best_score_     # highest accuracy
```

```python
import pandas as pd
results=pd.DataFrame(cls.cv_results_)
results
```

```python
results.head()
```

```python
gsr=results[['param_C','param_kernel','mean_test_score']]
gsr    #grid search result
```
