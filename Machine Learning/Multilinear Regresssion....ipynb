```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('50_Startups.csv')

# Extract the independent variables (X) and dependent variable (y)
X = df[['R&D Spend', 'Administration', 'Marketing Spend']]
y = df['Profit']

# Create a design matrix
X = np.column_stack((X, np.ones(len(X))))

# Perform multilinear regression
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

# Plot the data and regression line
fig, ax = plt.subplots()
ax.scatter(df['R&D Spend'], y, label='R&D Spend')
ax.scatter(df['Administration'], y, label='Administration')
ax.scatter(df['Marketing Spend'], y, label='Marketing Spend')
ax.set_xlabel('Expenditure')
ax.set_ylabel('Profit')
ax.set_title('Startups Profit vs. Expenses')
x_range = np.linspace(df['R&D Spend'].min(), df['R&D Spend'].max(), 100)
y_range = np.linspace(df['Administration'].min(), df['Administration'].max(), 100)
x_mesh, y_mesh = np.meshgrid(x_range, y_range)
z_mesh = beta_hat[0]*x_mesh + beta_hat[1]*y_mesh + beta_hat[2]
ax.contour(x_mesh, y_mesh, z_mesh, levels=10, colors='red')
ax.legend()
plt.show()
```
