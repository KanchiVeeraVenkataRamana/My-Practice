```python
from sklearn.datasets import load_iris
#dir(datasets)
df=load_iris()
df
```

```python
import pandas as pd
x=pd.DataFrame(data=df.data,columns=df.feature_names)
y=pd.Series(data=df.target)
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train.shape)
print(x_train.size)
x_train.head()
```

```python
print(x_test.shape)
print(x_test.size)
x_test.head()
```

```python
print(y_train.shape)
print(y_train.size)
y_train.head()
```

```python
print(y_test.shape)
print(y_test.size)
y_test.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense
ml=Sequential()
ml.add(Dense(60,input_dim=4,activation='relu'))
ml.add(Dense(45,activation='softmax'))
ml.add(Dense(30,activation='relu'))
ml.add(Dense(10,activation='linear'))
ml.add(Dense(1,activation='sigmoid'))
ml.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
ml.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
ml.evaluate(x_test,y_test)
```

```python
ml.predict(x_test)
```

```python
from keras.layers import Dropout
ml1=Sequential()
ml1.add(Dense(60,input_dim=4,activation='relu'))
ml1.add(Dropout(0.5))
ml1.add(Dense(45,activation='softmax'))
ml1.add(Dense(30,activation='relu'))
ml1.add(Dense(10,activation='linear'))
ml1.add(Dense(1,activation='sigmoid'))
ml1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
ml1.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
ml1.evaluate(x_train,y_train)
```

```python
ml2=Sequential()
ml2.add(Dense(60,input_dim=4,activation='relu'))
ml2.add(Dropout(0.5))
ml2.add(Dense(45,activation='softmax'))
ml2.add(Dropout(0.7))
ml2.add(Dense(30,activation='relu'))
ml2.add(Dense(10,activation='linear'))
ml2.add(Dense(1,activation='sigmoid'))
ml2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
ml2.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
ml2.evaluate(x_train,y_train)
```

```python
ml3=Sequential()
ml3.add(Dense(60,input_dim=4,activation='relu'))
ml3.add(Dropout(0.5))
ml3.add(Dense(45,activation='softmax'))
ml3.add(Dropout(0.7))
ml3.add(Dense(30,activation='relu'))
ml3.add(Dense(10,activation='linear'))
ml3.add(Dropout(0.5))
ml3.add(Dense(1,activation='sigmoid'))
ml3.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
ml3.fit(x_train,y_train,epochs=100,batch_size=8)
```
