```python
import pandas as pd
df=pd.read_csv('sonar_dataset.csv')
df.head()
```

```python
df.isna().sum().sum()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1:]
x.head()
```

```python
y.sample(5)
```

```python
y.value_counts()
```

```python
y=pd.get_dummies(y,drop_first=True)
y.sample(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train.shape)
print(x_train.size)
x_train.head()
```

```python
print(x_test.shape)
print(x_test.size)
x_test.head()
```

```python
print(y_train.shape)
print(y_train.size)
y_train.head()
```

```python
print(y_test.shape)
print(y_test.size)
y_test.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
#1.Binar cross entropy
ml=Sequential()
ml.add(Dense(60,input_dim=60,activation='relu'))
ml.add(Dense(30,activation='relu'))
ml.add(Dense(15,activation='softmax'))
ml.add(Dense(1,activation='sigmoid'))
ml.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy','mae'])
ml.fit(x_train,y_train,epochs=50,batch_size=8)
```

```python
ml.evaluate(x_test,y_test)
```

```python
loss,accuracy,mae=ml.evaluate(x_test,y_test)
print(f'Loss:{loss}, Accuracy:{accuracy*100}, mean absolute error:{mae}')
```

```python
ml.predict(x_test)
```

```python
# 2.Mean Squared Error
ml1=Sequential()
ml1.add(Dense(60,input_dim=60,activation='relu'))
ml1.add(Dense(15,activation='softmax'))
ml1.add(Dense(1,activation='linear'))
ml1.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy','mae'])
ml1.fit(x_train,y_train,epochs=50,batch_size=8)
```

```python
ml1.evaluate(x_test,y_test)
```

```python
import warnings
warnings.filterwarnings('ignore')

# 3.categorical cross entropy
ml2=Sequential()
ml2.add(Dense(60,input_dim=60,activation='relu'))
ml2.add(Dense(15,activation='softmax'))
ml2.add(Dense(1,activation='linear'))
ml2.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy','mae'])
ml2.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
ml2.evaluate(x_test,y_test)
```

```python
# 4.Kullback-Leibler Divergence (KL Divergence) 
from tensorflow import keras
ml3=Sequential()
ml3.add(Dense(60,input_dim=60,activation='relu'))
ml3.add(Dense(15,activation='relu'))
ml3.add(Dense(1,activation='softmax'))
ml3.compile(loss=keras.losses.KLDivergence(),optimizer='adam',metrics=['accuracy','mae'])
ml3.fit(x_train,y_train,epochs=100,batch_size=8,validation_split=0.2)
```

```python
ml3.evaluate(x_test,y_test)
```

```python
import tensorflow as tf
ml4=Sequential()
ml4.add(Dense(60,input_dim=60,activation='relu'))
ml4.add(Dense(15,activation='relu'))
ml4.add(Dense(1,activation='softmax'))
ml4.compile(loss=tf.keras.losses.Huber(delta=1.0),optimizer='adam',metrics=['accuracy','mae'])
ml4.fit(x_train,y_train,epochs=100,batch_size=8,validation_split=0.2)
```

```python
ml4.evaluate(x_test,y_test)
```
