```python
from sklearn.datasets import load_wine
#dir(datasets)
df=load_wine()
df
```

```python
import pandas as pd
x=pd.DataFrame(data=df.data,columns=df.feature_names)
y=pd.Series(data=df.target)
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_train.size)
x_train.head()
```

```python
print(x_test.shape)
print(x_test.size)
x_test.head()
```

```python
print(y_train.shape)
print(y_train.size)
y_train.head()
```

```python
print(y_test.shape)
print(y_test.size)
y_test.head()
```

```python
# 1.Stochastic Gradient Descent (SGD)
from keras.models import Sequential
from keras.layers import Dense
ml=Sequential()
ml.add(Dense(30,input_dim=13,activation='relu'))
ml.add(Dense(15,activation='softmax'))
ml.add(Dense(1,activation='sigmoid'))
ml.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
ml.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
ml.evaluate(x_test,y_test)
```

```python
# 2.Adam Optimizer
from keras.optimizers import SGD
sgd = SGD(learning_rate=0.01, momentum=0.9)
#model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
ml1=Sequential()
ml1.add(Dense(30,input_dim=13,activation='relu'))
ml1.add(Dense(15,activation='relu'))
ml1.add(Dense(1,activation='sigmoid'))
ml1.compile(loss='binary_crossentropy',optimizer=sgd,metrics=['accuracy'])
ml1.fit(x_train,y_train,epochs=150,batch_size=10,verbose=0)
```

```python
import warnings
warnings.filterwarnings('ignore')

ml1.evaluate(x_test,y_test)
```

```python
# 3.Adagrad Optimizer
import tensorflow as tf
ml2=Sequential()
ml2.add(Dense(30,input_dim=13,activation='relu'))
ml2.add(Dense(15,activation='relu'))
ml2.add(Dense(1,activation='sigmoid'))
ml2.compile(loss='binary_crossentropy',optimizer=tf.keras.optimizers.Adagrad(learning_rate=0.01),metrics=['accuracy'])
ml2.fit(x_train,y_train,epochs=150,batch_size=32)
```

```python
ml2.evaluate(x_test,y_test)
```

```python
# 4.Adadelta Optimizer
ml3=Sequential()
ml3.add(Dense(30,input_dim=13,activation='relu'))
ml3.add(Dense(15,activation='relu'))
ml3.add(Dense(1,activation='sigmoid'))
ml3.compile(loss='binary_crossentropy',optimizer='adadelta',metrics=['mae','accuracy'])
ml3.fit(x_train,y_train,epochs=150,batch_size=32,verbose=1)
```

```python
ml3.evaluate(x_test,y_test)
```

```python
# 5.RMSprop Optimizer
import keras
ml4=Sequential()
ml4.add(Dense(30,input_dim=13,activation='relu'))
ml4.add(Dense(15,activation='relu'))
ml4.add(Dense(1,activation='sigmoid'))
ml4.compile(loss='binary_crossentropy',optimizer=keras.optimizers.RMSprop(learning_rate=0.001),metrics=['accuracy','mae'])
ml4.fit(x_train,y_train,epochs=150,batch_size=32)
```

```python
ml4.evaluate(x_test,y_test)
```

```python
# 6.Momentum Optimizer
from tensorflow.keras.optimizers import SGD
momentum_optimizer = SGD(lr=0.01, momentum=0.9)
ml5=Sequential()
ml5.add(Dense(30,input_dim=13,activation='relu'))
ml5.add(Dense(15,activation='relu'))
ml5.add(Dense(1,activation='sigmoid'))
ml5.compile(loss='mean_squared_error', optimizer=momentum_optimizer,metrics=['accuracy','mae'])
ml5.fit(x_train,y_train,epochs=150,batch_size=16)
```

```python
ml5.evaluate(x_test,y_test)
```

```python
# 7.Nesterov Accelerated Gradient (NAG) Optimizer
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True)
ml6=Sequential()
ml6.add(Dense(30,input_dim=13,activation='relu'))
ml6.add(Dense(15,activation='relu'))
ml6.add(Dense(1,activation='sigmoid'))
ml6.compile(loss='mean_squared_error', optimizer=optimizer,metrics=['accuracy','mae'])
ml6.fit(x_train,y_train,epochs=150,batch_size=16)
```

```python
ml6.evaluate(x_test,y_test)
```
