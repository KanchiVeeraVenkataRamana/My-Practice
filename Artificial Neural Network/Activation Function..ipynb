```python
from sklearn.datasets import load_wine
#dir(datasets)
df=load_wine()
df
```

```python
import pandas as pd
x=pd.DataFrame(data=df.data, columns=df.feature_names)
y=pd.Series(data=df.target)
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print(x_train.shape)
print(x_train.size)
x_train.head()
```

```python
print(x_test.shape)
print(x_test.size)
x_test.head()
```

```python
print(y_train.shape)
print(y_train.size)
y_train.head()
```

```python
print(y_test.shape)
print(y_test.size)
y_test.head()
```

```python
# 1.Sigmoid
from keras.models import Sequential
from keras.layers import Dense
ml=Sequential()
ml.add(Dense(15,input_dim=13,activation='relu'))
ml.add(Dense(1,activation='sigmoid'))
ml.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy','mae'])
ml.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
ml.evaluate(x_test,y_test)
```

```python
# 2.tanh

ml1=Sequential()
ml1.add(Dense(15,input_dim=13,activation='relu'))
ml1.add(Dense(1,activation='tanh'))
ml1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy','mae'])
ml1.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
import warnings
warnings.filterwarnings('ignore')

ml1.evaluate(x_test,y_test)
```

```python
# 3.linear
ml2=Sequential()
ml2.add(Dense(15,input_dim=13,activation='relu'))
ml2.add(Dense(1,activation='linear'))
ml2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
ml2.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
import warnings
warnings.filterwarnings('ignore')

ml2.evaluate(x_test,y_test)
```

```python
# 4.softmax
ml3=Sequential()
ml3.add(Dense(15,input_dim=13,activation='relu'))
ml3.add(Dense(1,activation='softmax'))
ml3.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
ml3.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
ml3.evaluate(x_test,y_test)
```
