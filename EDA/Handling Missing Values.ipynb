# 1.Fillna Method

```python
import pandas as pd
df=pd.read_csv('Placement_Data_Full_Class.csv')
df
```

```python
#fillna mean method
df=pd.read_csv('Placement_Data_Full_Class.csv')
df.head()
```

# 3.Replace Method

```python
import numpy as np
import pandas as pd
df=pd.read_csv('Placement_Data_Full_Class.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df['salary'].replace(np.NaN,df['salary'].mean(),inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df['salary'].replace(np.NaN,df['salary'].median(),inplace=True)
df.head()
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv('Placement_Data_Full_Class.csv')
df.head()
```

```python
df['salary'].replace(np.NaN,df['salary'].mode,inplace=True)
df.head()
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv('Placement_Data_Full_Class.csv')
df.head()
```

```python
df['salary'].replace(to_replace=np.NaN,value=-9999,inplace=True)
df.head()
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv('Placement_Data_Full_Class.csv')
df.head()
```

```python
df['salary'].replace(to_replace=np.NaN,value='NO DATA',inplace=True)
df.head()
```
