# Outliers Detection

```python
import pandas as pd
df2=pd.read_csv('weight-height.csv')
df2
```

```python
df2.head()
```

```python
df2.describe(include='all')
```

```python
df2.describe()
```

```python
#quantiles Q1 and Q3
Q1=df2.Height.quantile(0.25)
Q3=df2.Height.quantile(0.75)
Q1,Q3
```

```python
#IQR=Q3-Q1
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit,upper_limit
```

```python
import seaborn as sn
sn.boxplot(df2['Height'])
```

```python
sn.boxplot(df2['Weight'])
```

```python
outlier_data=df2[(df2.Height<lower_limit)|(df2.Height>upper_limit)]
outlier_data
```

```python
no_outliers_data=df2[(df2.Height>lower_limit)&(df2.Height<upper_limit)]
no_outliers_data
```

```python
no_outliers_data.size
```

```python
outlier_data.size
```

```python
df2.shape[0] - no_outliers_data.shape[0]
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sn.boxplot(df2['Height'])
```

```python
#percentile method
df2['Height']=df2['Height'].clip(lower=df2['Height'].quantile(0.25), upper=df2['Height'].quantile(0.75))
df2['Height']
```

```python
sn.boxplot(df2['Height'])
```

```python
sn.boxplot(df2['Weight'])
```

```python
df2['Weight']=df2['Weight'].clip(lower=df2['Weight'].quantile(0.25),upper=df2['Weight'].quantile(0.75))
df2['Weight']
```

```python
sn.boxplot(df2['Weight'])
```

# Simple Imputer

```python
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
df2=pd.DataFrame()
df2['Col']=[75,88,np.nan,94,168,np.nan,543]
df2
```

```python
#mean
s_mean=SimpleImputer(strategy='mean')
mean=s_mean.fit_transform(df2)
mean
```

```python
#median is even
s_median=SimpleImputer(strategy='median')
median=s_median.fit_transform(df2)
median
```

```python
#median is odd
df3=pd.DataFrame()
df3['Col']=[75,88,np.nan,94,168,220,543]
s_median=SimpleImputer(strategy='median')
median=s_median.fit_transform(df3)
median
```

```python
#mode
s_mode=SimpleImputer(strategy='most_frequent')
mode=s_mode.fit_transform(df2)
mode
```

```python
#constant
s_constant=SimpleImputer(strategy='constant',fill_value=100)
constant=s_constant.fit_transform(df2)
constant
```
