```python
import matplotlib.pyplot as pt
import seaborn as sn
correlation_metrics=data.corr()
fig=pt.figure(figsize=(10,9))
sn.heatmap(correlation_metrics,annot=True)
pt.show()
print(correlation_metrics)
```

```python
#no.of people with no diabets
no_sugar=0
result=data['Outcome']
for i in range(len(result)):
    if result[i]==0:
        no_sugar=no_sugar+1
print(no_sugar)
```

```python
#no.of people without diabets
sugar=0
for i in range(len(result)):
    if result[i]==1:
        sugar=sugar+1
print(sugar)
```

```python
#plotting count plot with seaborn
sn.countplot(x=data.Outcome)
pt.title('No diabetes vs Diabetes')
```

```python
from sklearn.model_selection import train_test_split
x=data.drop(['Outcome'],axis=1)
y=data['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
l=LogisticRegression()
l.fit(x_train,y_train)
```

```python
y_pred=l.predict(x_test)
y_pred
```

```python
from sklearn import metrics
cm=metrics.confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
accuracy
```
