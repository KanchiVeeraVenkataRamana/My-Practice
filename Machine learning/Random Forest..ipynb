```python
from sklearn import datasets
dir(datasets)
```

```python
data=datasets.load_wine()
data
```

```python
import pandas as pd
x=pd.DataFrame(data=data.data,columns=data.feature_names)
y=pd.Series(data=data.target)
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
print(x_train.shape)
print(x_train.size)
```

```python
print(x_test.shape)
print(x_test.size)
```

```python
print(y_test.shape)
print(y_test.size)
```

```python
print(y_train.shape)
print(y_train.size)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=10,criterion='entropy',max_depth=2,random_state=12)
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn import metrics
import seaborn as sn
cm=metrics.confusion_matrix(y_test,y_pred)
sn.heatmap(cm,annot=True)
```

```python
print(metrics.classification_report(y_pred,y_test))
```

```python
metrics.accuracy_score(y_pred,y_test)
```

```python
metrics.precision_score(y_pred,y_test,average='macro')
```

```python
metrics.f1_score(y_pred,y_test,average='micro')
```

```python
metrics.recall_score(y_pred,y_test,average='macro')
```

```python
metrics.mean_squared_error(y_pred,y_test)
```

```python
metrics.mean_absolute_error(y_pred,y_test)
```

```python
metrics.r2_score(y_pred,y_test)
```

```python
#root mean square error
import math
import numpy as np
MSE=metrics.mean_squared_error(y_test,y_pred)
RMSE=math.sqrt(MSE)
RMSE
```

```python
#Adjusted R2 score
import numpy as np
r2_score=metrics.r2_score(y_test,y_pred)
n=y_test.shape[0]
p=3
x=(1-r2_score)
y=(n-1)/(n-p-1)
adj_r2=(1-(x*y))
print('Adjusted_R2:',adj_r2)
```
