```python
import pandas as pd
from sklearn import datasets
df=datasets.load_breast_cancer()
df
```

```python
x=pd.DataFrame(df.data,columns=df.feature_names)
y=pd.Series(df.target)
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_train.size)
x_train.head()
```

```python
print(x_test.shape)
print(x_test.size)
x_test.head()
```

```python
print(y_train.shape)
print(y_train.size)
y_train.head()
```

```python
print(y_test.shape)
print(y_test.size)
y_test.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion='entropy')
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn import metrics
import seaborn as sn
cm=metrics.confusion_matrix(y_pred,y_test)
sn.heatmap(cm,annot=True)
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as pt
from sklearn import tree
clf=tree.DecisionTreeClassifier(criterion='entropy',random_state=0)
cls=clf.fit(x,y)
pt.figure(figsize=(14,10))
tree.plot_tree(cls,label='all',rounded=True,node_ids=True,precision=3,fontsize=10)  #max_depth=2
```

```python
import matplotlib.pyplot as pt
from sklearn import tree
clf=tree.DecisionTreeClassifier(criterion='entropy',random_state=0)
cls=clf.fit(x,y)
pt.figure(figsize=(23,10))
tree.plot_tree(clf,node_ids=True,feature_names=df.feature_names,class_names=df.target_names,filled=True,rounded=True,fontsize=8)   #max_depth=2
```
