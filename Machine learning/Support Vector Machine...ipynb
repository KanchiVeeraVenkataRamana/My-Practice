```python
from sklearn import datasets
df=datasets.load_iris()
df
```

```python
df.feature_names
```

```python
x=df.data
x
```

```python
y=df.target
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
from sklearn.svm import SVC
cls=SVC(kernel='linear')
cls.fit(x_train,y_train)
```

```python
cls.score(x_test,y_test)
```

```python
y_pred=cls.predict(x_test)
y_pred
```

```python
import warnings
warnings.filterwarnings('ignore')


from sklearn import metrics
print(metrics.classification_report(y_pred,y_test))
```

```python
metrics.confusion_matrix(y_pred,y_test)
```

```python
import matplotlib.pyplot as pt
cm=metrics.confusion_matrix(y_pred,y_test)
cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[0,1,2])
cm_display.plot()
pt.show()
```

```python
metrics.accuracy_score(y_pred,y_test)
```

```python
metrics.precision_score(y_pred,y_test,average='macro')
```

```python
metrics.recall_score(y_pred,y_test,average='micro')
```

```python
metrics.f1_score(y_pred,y_test,average='macro')
```

```python
metrics.mean_squared_error(y_pred,y_test)
```

```python
metrics.mean_absolute_error(y_pred,y_test)
```

```python
metrics.r2_score(y_pred,y_test)
```

```python
#root mean square error
import math
import numpy as np
MSE=metrics.mean_squared_error(y_test,y_pred)
RMSE=math.sqrt(MSE)
RMSE
```

```python
#Adjusted R2 score
import numpy as np
r2_score=metrics.r2_score(y_test,y_pred)
n=y_test.shape[0]
p=3
x=(1-r2_score)
y=(n-1)/(n-p-1)
adj_r2=(1-(x*y))
print('Adjusted_R2:',adj_r2)
```
