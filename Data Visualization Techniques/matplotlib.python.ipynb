```python
import matplotlib.pyplot as pt
x=[0,1,2,3,4]
y1=[20,10,70,30,60]
y2=[68,29,10,69,20]
y3=[50,20,40,10,49]
y4=[80,30,20,10,49]
y5=[79,20,40,20,10]
y6=[90,30,20,40,20]
pt.suptitle('MULTIPLE PLOTS')
print(x,y1)
pt.subplot(3,2,1)  #row,column,index
pt.plot(x,y1,marker='.',ls='dashed')
pt.title('plot-1')
pt.subplot(3,2,2)
print(x,y2)
pt.plot(x,y2,marker='D',ls='dotted')
pt.title('plot-2')
pt.subplot(3,2,3)
print(x,y3)
pt.plot(x,y3,marker='o',ls='dashdot')
pt.title('plot-3')
pt.subplot(3,2,4)
print(x,y4)
pt.plot(x,y4,marker='*',ls='none')
pt.title('plot-4')
pt.subplot(3,2,5)
print(x,y5)
pt.plot(x,y5,marker='P')
pt.title('plot-5')
pt.subplot(3,2,6)
print(x,y6)
pt.plot(x,y6,marker='P')
pt.title('plot-6')
pt.show()
```

```python
import matplotlib.pyplot as pt
x=[0,1,2,3,4]
y1=[20,10,70,30,60]
y2=[68,29,10,69,20]
y3=[50,20,40,10,49]
y4=[80,30,20,10,49]
y5=[79,20,40,20,10]
y6=[90,30,20,40,20]
pt.suptitle('MULTIPLE PLOTS')
print(x,y1)
pt.subplot(3,3,1)  #row,column,index
pt.plot(x,y1,marker='.',ls='dashed')
pt.title('plot-1')
pt.subplot(3,3,2)
print(x,y2)
pt.plot(x,y2,marker='D',ls='dotted')
pt.title('plot-2')
pt.subplot(3,3,3)
print(x,y3)
pt.plot(x,y3,marker='o',ls='dashdot')
pt.title('plot-3')
pt.subplot(3,3,4)
print(x,y4)
pt.plot(x,y4,marker='*',ls='none')
pt.title('plot-4')
pt.subplot(3,3,5)
print(x,y5)
pt.plot(x,y5,marker='P')
pt.title('plot-5')
pt.subplot(3,3,6)
print(x,y6)
pt.plot(x,y6,marker='^')
pt.title('plot-6')
pt.show()
```

```python
#xticks(),yticks()
x=[10,20,30,40,50,60]
y=[100,200,300,400,500,600]
pt.plot(x,y,marker='*',ms=13,ls='dashdot')
pt.xlabel('x-axis')
pt.ylabel('y-axis')
pt.title('xticks(),yticks()')
pt.xticks(x,['A','B','C','D','E','F'])
pt.yticks(y,['0-100','100-200','200-300','300-400','400-500','500-600'])
pt.show()
```

```python
#legend
x=[10,20,30,40,50,60]
y=[100,200,300,400,500,600]
pt.plot(x,marker='*',ms=13,ls='dashdot',c='y',label='line1')
pt.plot(y,marker='^',ms=10,ls='dashdot',c='r',label='line2')
pt.legend(loc='upper left',framealpha=1,facecolor='green',edgecolor='yellow',fancybox=True,shadow=True,fontsize=10) 
pt.show()
```

```python
#multiple bar graphs
import numpy as np
import matplotlib.pyplot as pt
groups=['G1','G2','G3','G4','G5']
men=[22,30,35,35,26]
women=[25,32,30,35,29]
x=np.arange(len(groups))
w=0.33
pt.bar(x+w/2,men,width=w,color='red',label='Men')
pt.bar(x-w/2,women,width=w,color='green',label='Women')
pt.xticks(x,groups)
pt.legend(loc='upper right',edgecolor='black')
pt.show()
```

```python
import matplotlib.pyplot as pt
import numpy as np
labels=['G1','G2','G3','G4','G5']
men_means = (20, 35, 30, 35, 27)
women_means = (25, 32, 34, 20, 25)
x=np.arange(len(labels))
w=0.3
a=pt.bar(x-w/2,men_means,width=w,color='red',label='Men')
b=pt.bar(x+w/2,women_means,width=w,color='green',label='Women')
pt.xticks(x,labels)
pt.legend(edgecolor='black')
pt.bar_label(a,padding=3)
pt.bar_label(b,padding=3)
pt.tight_layout()
pt.show()
```

# Scatter Plot

```python
x=[10,20,30,40,50]
y=[200,300,100,500,400]
colors=['red','green','blue','pink','black']
sizes=[100,50,300,250,150]
pt.scatter(x,y,c=colors,s=sizes)
x1=[20,30,40,50,60]
y1=[300,200,29,10,489]
pt.scatter(x1,y1,c=colors,s=sizes)
pt.show()
```

```python
#cmap and colorbar
pt.scatter(x,y,c=colors,s=sizes,cmap='rainbow',alpha=0.5)
pt.colorbar()
pt.show()
```
