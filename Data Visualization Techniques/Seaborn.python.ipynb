# Seaborn

```python
import pandas as pd
import seaborn as sn
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv('train.csv')
df
```

```python
df.head()
```

# Distribution plots

```python
#displot
sn.displot(x=df['Age'])
```

```python
sn.distplot(x=df['Age'],bins=10)
```

```python
sn.displot(x=df['Age'],bins=10,kde=False)
```

```python
#joint plot
sn.jointplot(x=df['Age'],y=df['Fare'],kind='scatter')
```

```python
sn.jointplot(x=df['Age'],y=df['Fare'],kind='hex')
```

```python
#pairplot
sn.pairplot(df)
```

```python
#Rugplot
sn.rugplot(x=df['Age'])
```

# Categorical Plots

```python
#Bar plot
sn.barplot(x=df['Age'],y=df['Fare'])
```

```python
#countplot
sn.countplot(df['Pclass'])
```

```python
#boxplot
sn.boxplot(x=df['Sex'],y=df['Age'])
```

```python
#violin plot
sn.violinplot(x=df['Sex'],y=df['Age'])
```

# Advanced Plots

```python
#strip plot
sn.stripplot(x=df['Pclass'],y=df['Age'])
```

```python
#swarm plot/catplot
sn.swarmplot(x=df['Pclass'],y=df['Age'])
```

```python
#point plot
sn.pointplot(x=df['Pclass'],y=df['Age'])
```

# Matrix Plots

```python
#heatmap
import matplotlib.pyplot as pt
pt.figure(figsize=(10,8))
sn.heatmap(df.corr(),annot=True,cmap='viridis')
```

```python
sn.heatmap(df.isnull(),cmap='viridis')
```

```python
sn.heatmap(df.isnull(),cmap='viridis',yticklabels=False,cbar=False)
```

```python
sn.heatmap(df.isna(),cmap='viridis',cbar=False,yticklabels=False)
```

```python
#Cluster Map
sn.clustermap(df.corr(),annot=True,cmap='viridis')
```

# Grids

```python
#Facet Grids
sn.FacetGrid(df,col='Pclass',row='Sex').map(sn.distplot,'Age')
```

```python
#Regression Plot
sn.lmplot(x='Age',y='PassengerId',data=df,hue='Sex')
```

```python
sn.regplot(x='Age',y='PassengerId',data=df,)
```
