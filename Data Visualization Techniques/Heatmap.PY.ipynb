```python
#data applying
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sn
a=np.random.randint(low=1,high=100,size=(10,10))
print(a)
pt.figure(figsize=(10,5))
hm=sn.heatmap(data=a)
pt.show()
```

```python
# applying vmin,vmax
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sn
a=np.random.randint(low=1,high=100,size=(10,10))
print(a)
hm=sn.heatmap(data=a,vmin=40,vmax=70)
pt.show()
```

```python
#choosing the colormap
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sn
a=np.random.randint(low=1,high=100,size=(10,10))
print(a)
hm=sn.heatmap(data=a,cmap='tab20')
pt.show()
```

```python
#centering the heatmap
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sn
a=np.random.randint(low=1,high=100,size=(10,10))
print(a)
hm=sn.heatmap(data=a,cmap='tab20',center=20)
pt.show()
```

```python
#displaying the cell values
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sn
a=np.random.randint(low=1,high=100,size=(10,10))
print(a)
hm=sn.heatmap(data=a,cmap='tab20',annot=True)
pt.show()
```

```python
#customizing the seperate lines
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sn
a=np.random.randint(low=1,high=100,size=(10,10))
print(a)
#hm=sn.heatmap(data=a,cmap='tab20',annot=True,linewidth=7,linecolor='green')
hm=sn.heatmap(data=a,linewidth=2,linecolor='yellow')
pt.show()
```

```python
#removing the colorbar
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sn
a=np.random.randint(low=1,high=100,size=(10,10))
print(a)
hm=sn.heatmap(data=a,cbar=False)
pt.show()
```

```python
#Removing the labels
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sn
a=np.random.randint(low=1,high=100,size=(10,10))
print(a)
hm=sn.heatmap(data=a,xticklabels=False,yticklabels=False)
pt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sn
a=np.random.randint(low=1,high=100,size=(10,10))
print(a)
hm=sn.heatmap(data=a,vmin=20,vmax=80,cmap='tab20',center=0,annot=True,linewidth=2,linecolor='Yellow',cbar=False)
pt.show()
```
