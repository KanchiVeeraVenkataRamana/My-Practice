```python
import matplotlib.pyplot as pt
import numpy as np
x=np.array([45,55,67,88,90,65,60,78])
y=np.array([55,51,77,98,70,75,82,82])

font1={'family':'serif','color':'b','size':18}
font2={'family':'serif','color':'r','size':14}
pt.plot(x,y)
pt.title("Maths and English Marks",fontdict=font1,loc='left')
pt.xlabel("Marks in Maths",fontdict=font2)
pt.ylabel("Marks in English",fontdict=font2)
```

```python
import matplotlib.pyplot as pt
import numpy as np
x=np.array([45,55,67,88,90,65,60,78])
y=np.array([55,51,77,98,70,75,82,82])

font1={'family':'serif','color':'b','size':18}
font2={'family':'serif','color':'r','size':14}
pt.plot(x,y)
pt.grid(axis='y')
pt.title("Maths and English Marks",fontdict=font1,loc='left')
pt.xlabel("Marks in Maths",fontdict=font2)
pt.ylabel("Marks in English",fontdict=font2)
```

```python
import matplotlib.pyplot as pt
import numpy as np
x=np.array([45,55,67,88,90,65,60,78])
y=np.array([55,51,77,98,70,75,82,82])

font1={'family':'serif','color':'b','size':18}
font2={'family':'serif','color':'r','size':14}
pt.plot(x,y)
pt.grid()
pt.title("Maths and English Marks",fontdict=font1,loc='left')
pt.xlabel("Marks in Maths",fontdict=font2)
pt.ylabel("Marks in English",fontdict=font2)
pt.show()
```

```python
import matplotlib.pyplot as pt
import numpy as np
#labels of x, and y axis
x1=np.array([45,55,67,88,90,65,60,78])
y1=np.array([55,51,77,98,70,75,81,82])
#row 1,column 2, subplot is 1
pt.subplot(3,1,1)
pt.plot(x1,y1)

x2=np.array([55,59,62,72,79,82,99,98])
y2=np.array([28,34,92,43,84,18,45,82])
#row1,columns2,subplot is 1
pt.subplot(3,1,2)
pt.plot(x2,y2)

x3=np.array([13,28,48,29,49,29,13,25])
y3=np.array([22,45,83,13,23,95,39,92])
#row1,column2,subplot is 1
pt.subplot(3,1,3)
pt.plot(x3,y3)
```

```python
#scattered plot 
import matplotlib.pyplot as pt
import numpy as np
#labels of x, and y axis
x1=np.array([45,55,56,23,78,14,78])
y1=np.array([23,46,36,29,86,48,59])
pt.scatter(x1,y1,c='#ff00ff',cmap='nipy_spectral')

x2=np.array([16,34,27,56,49,69,45])
y2=np.array([34,85,28,84,29,84,28])
pt.scatter(x2,y2,c='#ff8000',cmap='nipy_spectral')
```
