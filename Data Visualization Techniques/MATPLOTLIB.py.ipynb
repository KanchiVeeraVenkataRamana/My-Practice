SUBPLOTS/SUBPLOTS FUNCTIONS

```python
import matplotlib.pyplot as pt
x=np.arange(3,30,3)
y1=x*2
y2=x*3
y3=x*4
y4=x*5
pt.suptitle('MULTIPLE PLOTS')
print(x,y1)
pt.subplot(2,2,1)
pt.plot(x,y1,marker='.',ls='dashed')
pt.title('plot-1')
pt.subplot(2,2,2)
print(x,y2)
pt.plot(x,y2,marker='D',ls='dotted')
pt.title('plot-2')
pt.subplot(2,2,3)
print(x,y3)
pt.plot(x,y3,marker='o',ls='dashdot')
pt.title('plot-3')
pt.subplot(2,2,4)
print(x,y4)
pt.plot(x,y4,marker='*',ls='none')
pt.title('plot-4')
pt.show()
```

STACKED BAR GRAPH

```python
import matplotlib.pyplot as pt
years=['2017','2018','2019','2020']
placements=[150,275,324,295]
cse=[71,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]
eee=[53,47,36,52]
ece_start=[cse[i]+it[i] for i in range(len(cse))]
eee_start=[ece_start[i]+ece[i] for i in range(len(ece))]
pt.title('PLACEMENTS')
pt.title('PLACEMENT COMPARSION USING STACKED BAR PLOT GRAPH')
pt.bar(years,cse,width=0.3,color='black',label='CSE')
pt.bar(years,it,bottom=cse,width=0.3,color='green',label='IT')
pt.bar(years,ece,bottom=ece_start,width=0.3,color='blue',label='ECE')
pt.bar(years,eee,bottom=eee_start,width=0.3,color='red',label='EEE')
pt.xlabel('placements')
pt.ylabel('years')
pt.legend()
pt.ylim(0,350)
pt.show()
```

```python
import matplotlib.pyplot as pt
years=['2017','2018','2019','2020']
placements=[150,275,324,295]
cse=[71,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]
eee=[53,47,36,52]
pt.title('PLACEMENTS')
ece_start=[cse[i]+it[i]for i in range(len(cse))]
eee_start=[ece_start[i]+ece[i] for i in range(len(ece))]
pt.title('PLACEMENT COMPARSION USING STACKED BAR PLOT GRAPH')
#pt.bar(years,placements,width=0.3)
pt.barh(years,cse,height=0.3,color='black',label='CSE')
pt.barh(years,it,left=cse,height=0.3,color='green',label='IT')
pt.barh(years,ece,left=ece_start,height=0.3,color='red',label='ECE')
pt.barh(years,eee,left=eee_start,height=0.3,color='blue',label='EEE')
pt.xlabel('placements')
pt.ylabel('years')
pt.legend()
pt.xlim(0,300)
pt.show()
```

GROUPED / MULTIPLE BAR CHART

```python
import matplotlib.pyplot as pt
import numpy as np
years=['2017','2018','2019','2020']
cse=[74,82,89,79]
it=[58,72,76,61]
ece=[67,74,57,41]
eee=[53,45,57,41]
w=0.1
pt.figure(figsize=(10,7))
cse_bar=np.arange(len(years))
it_bar=[i+w for i in cse_bar]
ece_bar=[i+w for i in it_bar]
eee_bar=[i+w for i in ece_bar]
print(cse_bar)
pt.bar(cse_bar,cse,width=w,label='CSE')
print(it_bar)
pt.bar(it_bar,it,width=w,label='IT')
print(ece_bar)
pt.bar(ece_bar,ece,width=w,label='ECE')
print(eee_bar)
pt.bar(eee_bar,eee,width=w,label='EEE')
pt.title("PLACEMENT COMPARISION")
pt.xticks(cse_bar+w+w/2,years)
pt.ylim(0,100)
pt.xlabel('YEARS')
pt.ylabel('PLACEMENT COUNT')
pt.legend()
pt.show()
```

```python
import matplotlib.pyplot as pt
import numpy as np
years=['2017','2018','2019','2020']
cse=[74,82,89,79]
it=[58,72,76,61]
ece=[67,74,57,41]
eee=[53,45,57,41]
h=0.1
pt.figure(figsize=(10,7))
cse_bar=np.arange(len(years))
it_bar=[i+w for i in cse_bar]
ece_bar=[i+w for i in it_bar]
eee_bar=[i+w for i in ece_bar]
print(cse_bar)
pt.barh(cse_bar,cse,height=h,label='CSE')
print(it_bar)
pt.barh(it_bar,it,height=h,label='IT')
print(ece_bar)
pt.barh(ece_bar,ece,height=h,label='ECE')
print(eee_bar)
pt.barh(eee_bar,eee,height=h,label='EEE')
pt.title("PLACEMENT COMPARISION")
pt.xticks(cse_bar+h+h/2,years)
pt.xlim(0,100)
pt.xlabel('YEARS')
pt.ylabel('PLACEMENT COUNT')
pt.legend()
pt.show()
```
