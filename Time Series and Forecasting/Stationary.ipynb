Statistical Test
1. Augmented Dickey-Fuller (ACF) test
2. Kwiatkowsi-phillips-schmidit-Shin (KPSS) test
3. Rolling Statistical test
4. Phillips-perron (PP) test

```python
# 1. Augmented Dickey_Fuller (ACF) test
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
df = pd.read_csv('/content/covid_19_india.csv')
df
```

```python
df.head()
```

```python
df.describe()
```

```python
result = sm.tsa.adfuller(df['Confirmed'])
result
```

```python
print('ADF Statistic: ',result[0])
print("p-value:",result[1])
print('Critical Values:')
for key, value in result[4].items():
  print("\t %s: %.3f"%(key,value))
```

```python
plt.plot(df['Confirmed'])
plt.title('COVID-19 Cases')
plt.xlabel('number of cases')
plt.ylabel('data')
plt.show()
```

```python
# 2. Kwiatkowsi-phillips-schmidit-Shin (KPSS) test
import warnings
warnings.filterwarnings('ignore')

result = sm.tsa.kpss(df['Confirmed'])
result
```

```python
print('KPSS Statistics: ',result[0])
print('P-Value: ',result[1])
print('Critical value:')
for key,value in result[3].items():
  print("\t %s%.3f"%(key,value))
```

```python
# 3. Random Statistics
roll_mean = df['Confirmed'].rolling(window = 5).mean()
roll_std = df['Confirmed'].rolling(window = 5).std()
roll_sum = df['Confirmed'].rolling(window = 5).sum()
pt.figure(figsize = (7,7))
pt.plot(df['Date'], df['Confirmed'], label='Original')
pt.plot(df['Date'], roll_mean, label='Rolling_mean')
pt.plot(df['Date'], roll_std, label='Rolling_std')
pt.plot(df['Date'], roll_sum, label='Rolling_sum')
pt.xlabel('Date')
pt.ylabel('Value')
pt.title('Rolling Statistics')
pt.legend()
pt.show()
```

```python
# 4. Phillips-perron (PP) test:
result = sm.tsa.adfuller(df['Confirmed'], autolag= 'AIC', regression='ct',maxlag=12)
result
```

```python
print('Test-Statistics: ',result[0])
print('P-Value: ',result[1])
print('Null-Hypothesis: ',result[2])
```
